```python
def format_time(time_int):
    # Convert the integer to string
    time_str = str(int(time_int))

    # Prepend zeros based on the length of the time string
    if len(time_str) == 1:
        time_str = '0' + time_str + '00'  # For 1-digit times, prepend '0' and append '00'
    elif len(time_str) == 2:
        time_str = '00' + time_str  # For 2-digit times, prepend '00'
    elif len(time_str) == 3:
        time_str = '0' + time_str   # For 3-digit times, prepend '0'

    # Check for '2400' and convert it to '00:00:00'
    if time_str == '2400':
        return '00:00:00'

    # Rearrange the string into HH:MM:SS format
    formatted_time = f"{time_str[:2]}:{time_str[2:4]}:00"
    
    return formatted_time

# List of your time columns
time_columns = ['DepTime', 'ArrTime', 'WheelsOff', 'WheelsOn']

# Applying the formatting to each time column
for col in time_columns:
    arr2019[col] = arr2019[col].apply(format_time)
```

```python
sw2019_raw['Route'] = sw2019_raw['OriginCityName'] + "-" + sw2019_raw['DestCityName']

route_count = sw2019_raw['Route'].value_counts().reset_index()
route_count_cutoff = 500
route_count_list = route_count[route_count['count'] >= route_count_cutoff].Route.tolist()

sw2019_raw = sw2019_raw[sw2019_raw['Route'].isin(route_count_list)]

# Now, we'll group by ROUTE and calculate average departure delay, average arrival delay, and cancellation rate
route_analysis = sw2019_raw.groupby('Route').agg({
    'DepDelayMinutes': 'mean',   # Average departure delay
    'ArrDelayMinutes': 'mean',   # Average arrival delay
    'Cancelled': 'mean'        # Cancellation rate
}).reset_index()

# Sort the results by average departure delay and cancellation rate for the top problematic routes
route_analysis_sorted = route_analysis.sort_values(by=['DepDelayMinutes', 'Cancelled'], ascending=[False, False])

route_analysis_sorted.head(10)
```
